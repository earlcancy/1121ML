```python
#1
import graphics
from graphics import *
from graphics import GraphWin

def main():
    # 創建一個視窗
    win = GraphWin("Face", 400, 400)

    # 畫頭部
    head = Circle(Point(200, 200), 100)
    head.setFill("moccasin")
    head.draw(win)

    # 畫眼睛
    eye1 = Circle(Point(150, 170), 20)
    eye1.setFill("black")
    eye1.draw(win)
    eye2 = Circle(Point(250, 170), 20)
    eye2.setFill("black")
    eye2.draw(win)

    # 畫鼻子
    nose = Circle(Point(200, 200), 5)
    nose.setFill("black")
    nose.draw(win)

    # 畫嘴巴
    mouth = Oval(Point(150, 240), Point(250, 260))
    mouth.setFill("black")
    mouth.draw(win)

    # 顯示視窗
    win.getMouse()
    win.close()

main()
```

```python
#2
import graphics
from graphics import *
from graphics import GraphWin

def main():
    # 創建一個視窗
    win = GraphWin("長方形", 400, 400)

    # 在視窗中顯示說明
    text1 = "請在視窗中點擊兩個點"
    message = Text(Point(200, 20), text1)
    message.draw(win)

    # 初始化變數
    p1 = None
    p2 = None

    # 等待使用者點擊兩個點
    while True:
        click = win.getMouse()

        if p1 is None:
            p1 = click
            continue
        else:
            p2 = click
            break

    # 繪製長方形
    x1, y1 = p1.getX(), p1.getY()
    x2, y2 = p2.getX(), p2.getY()
    w = x2 - x1
    h = y2 - y1
    rect = Rectangle(Point(x1, y1), Point(x2, y2))
    rect.setFill("khaki")
    rect.draw(win)

    # 計算邊界長及面積
    l = 2 * (w + h)
    a = w * h

    # 在視窗中顯示邊界長及面積
    text2 = "邊界長：" + str(abs(l)) + " 面積：" + str(abs(a))
    message.setText(text2)

    # 顯示視窗
    win.getMouse()
    win.close()
    
main()
```

```python
#3
import math
import graphics
from graphics import *
from graphics import GraphWin

def square(x):
    return x*x
# 計算兩個點之間的距離
def distance(p1, p2):
    dist = math.sqrt(square(p2.getX()-p1.getX())+square(p2.getY()-p1.getY()))
    return abs(dist)
def main():
    # 創建一個視窗
    win = GraphWin("長方形", 400, 400)

    # 在視窗中顯示說明
    text1 = "請在視窗中點擊三個點"
    message = Text(Point(200, 20), text1)
    message.draw(win)

    # 初始化變數
    p1 = None
    p2 = None
    p3 = None

    # 等待使用者點擊三個點
    while True:
        click = win.getMouse()
        click.draw(win)
        if p1 is None:
            p1 = click
            continue
        elif p2 is None:
            p2 = click
            continue
        else:
            p3 = click
            break

    # 繪製三角形
    triangle = Polygon(p1, p2, p3)
    triangle.setFill("teal")
    triangle.setOutline("cyan")
    triangle.draw(win)
    
    # 計算邊界長
    la = distance(p1 , p2)
    lb = distance(p2 , p3)
    lc = distance(p3 , p1)
    l = la + lb + lc
        
    # 計算面積
    s = (la + lb + lc) / 2
    a = math.sqrt(s * (s - la) * (s - lb) * (s - lc))
    
    # 在視窗中顯示邊界長及面積
    text2 = "邊界長：" + str(abs(int(l))) + " 面積：" + str(abs(int(a)))
    message.setText(text2)

    # 顯示視窗
    win.getMouse()
    win.close()

main()
```

```python
#4
# 回覆前n個自然數的總和
def sumN(n):
    sum = 0
    for i in range(1, n + 1):
        sum += i
    return sum
# 回覆前n個自然數的3次方總和
def sumNCubes(n):
    sum = 0
    for i in range(1, n + 1):
        sum += i ** 3
    return sum
def main():
    # 輸入自然數
    n = int(input("請輸一個自然數："))
    # 輸出總和
    print("自然數的總和：", sumN(n), "自然數的3次方總和：",sumNCubes(n))

main()
```

```python
#5
# 將陣列內的字串轉換為數值
def toNumbers(strList):
    for i in range(len(strList)):
        strList[i] = int(strList[i])

def main():
    strList = ["35", "200", "800", "360"]

    toNumbers(strList)
    sum = 0
    for i in range(len(strList)):
        sum += strList[i]
        
    # 輸出陣列
    print(strList)
    # 輸出總和
    print(sum)
    
main()
```

```python
#6
import graphics
from graphics import *
from graphics import GraphWin

def drawFace(center, size, win):
    # 取得滑鼠位置
    x1, y1 = center.getX(), center.getY()
    
    # 畫頭部
    head = Circle(center, size)
    head.setFill("moccasin")
    head.draw(win)

    # 畫眼睛
    eye1 = Circle(Point(x1-12.5, y1-7), 5)
    eye1.setFill("black")
    eye1.draw(win)
    eye2 = Circle(Point(x1+12, y1-7), 5)
    eye2.setFill("black")
    eye2.draw(win)

    # 畫鼻子
    nose = Circle(center, 1.5)
    nose.setFill("black")
    nose.draw(win)

    # 畫嘴巴
    mouth = Oval(Point(x1-12.5, y1+10), Point(x1+12.5, y1+20))
    mouth.setFill("red")
    mouth.setOutline("black")
    mouth.draw(win)

def main():
    # 創建一個視窗
    win = GraphWin("Face", 600, 500)

    # 在視窗中顯示說明
    text1 = "請在視窗中點擊"
    message = Text(Point(200, 20), text1)
    message.draw(win)
    
    # 在滑鼠點擊的位置畫臉
    while True:
        center = win.getMouse()
        drawFace(center, 25, win)
    
    # 顯示視窗
    win.getMouse()
    win.close()
    
main()
```
